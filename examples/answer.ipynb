## 문제 1

1부터 n까지의 합을 구하는 함수를 작성하시오.

```python
def sum_numbers(n):
    """1부터 n까지의 합을 반환합니다."""
    if n < 0:
        return 0
    return n * (n + 1) // 2

result = sum_numbers(10)
print(result)
```

## 문제 2

리스트에서 중복을 제거하고 정렬된 리스트를 반환하는 함수를 작성하시오.

```python
def remove_duplicates(lst):
    """리스트의 중복을 제거하고 정렬하여 반환합니다."""
    if not isinstance(lst, list):
        return []
    return sorted(set(lst))

test_list = [5, 2, 3, 2, 1, 4, 3]
result = remove_duplicates(test_list)
print(result)
```
