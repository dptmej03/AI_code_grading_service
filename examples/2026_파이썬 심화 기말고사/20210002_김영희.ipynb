## 문제 1

1부터 n까지의 합을 구하는 함수를 작성하시오.

```python
def calculate_sum(num):
    s = 0
    for x in range(1, num):
        s = s + x
    return s

answer = calculate_sum(10)
print(answer)
```

## 문제 2

리스트에서 중복을 제거하고 정렬된 리스트를 반환하는 함수를 작성하시오.

```python
def remove_dup(arr):
    new_list = []
    for item in arr:
        if item not in new_list:
            new_list.append(item)
    new_list.sort()
    return new_list

test = [5, 2, 3, 2, 1, 4, 3]
result = remove_dup(test)
print(result)
```
